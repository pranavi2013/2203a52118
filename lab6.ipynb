```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
a=pd.read_csv('/content/Japan_life_expectancy.csv')
```

```python
print(a)
```

```python
df= a.drop('Prefecture', axis=1)
print(df)
```

```python
df.head()
```

```python
df.columns
```

```python
# Separate features and target variable
x= df.drop('Welfare_exp', axis=1)
y= df['Welfare_exp']
```

```python
print(x)
```

```python
print(y)
```

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Create and fit the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict on the test set
y_pred = model.predict(x_test)
y_pred2 = model.predict(x_train)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
mse2 = mean_squared_error(y_train, y_pred2)

print('Mean Squared Error:', mse)
print('Mean Squared Error:', mse2)
```

```python
from sklearn.linear_model import Ridge

# Perform Ridge Regression
ridge = Ridge(alpha=0.05)  # You can adjust the alpha parameter for regularization strength
ridge.fit(x_train, y_train)

# Predict on the testing set
y_pred = ridge.predict(x_test)
y_pred2 = ridge.predict(x_train)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mse2 = mean_squared_error(y_train, y_pred2)

print('Mean Squared Error:', mse)
print('Mean Squared Error:', mse2)

# Optionally, you can plot the predicted vs actual values
import matplotlib.pyplot as plt
print("test data")
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()

print("train data")
plt.scatter(y_train, y_pred2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

```python
from sklearn.linear_model import Lasso
alpha = 0.05
lasso_model = Lasso(alpha=alpha)
lasso_model.fit(x_train, y_train)

y_pred = lasso_model.predict(x_test)
y_pred2 = lasso_model.predict(x_train)

mse = mean_squared_error(y_test, y_pred)
mse2 = mean_squared_error(y_train, y_pred2)

print("Mean Squared Error:", mse)
print("Mean Squared Error:", mse2)


# Optionally, you can plot the predicted vs actual values
import matplotlib.pyplot as plt
print("test data")
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()

print("train data")
plt.scatter(y_train, y_pred2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
